### convert the datatype of 'Date' column to date_time format

```python
df['date']=pd.to_datetime(df.date)
```

### add new column "year" the dataframe which contain year only

```python
df['year']=df.date.dt.year
```

### add new month  in the dataframe

```python
df['month']=df.date.dt.month
```

### all records where "no of crimes" is 0

### the max and min of year in england

```python
df.groupby('year').agg(['max','min'])
```

### find the max and min  no of crimes recorded per area

```python
group_area=df.groupby('area')
```
